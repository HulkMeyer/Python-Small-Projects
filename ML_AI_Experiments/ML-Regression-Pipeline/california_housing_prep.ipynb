# Loading the different Sheets

```python
#Import Data
long_lat = pd.read_csv('long_lat.csv')
cal_housing_low = pd.read_csv('cal_housing_low.csv')
cal_housing_med = pd.read_csv('cal_housing_medium.csv')
cal_housing_high = pd.read_csv('cal_housing_high.csv')
ocean_proximity = pd.read_csv('ocean_proximity.csv')
```

# Merge all 5 sheets together

```python
housing1 = long_lat.merge(cal_housing_low) #merge long_lat with cal_housing_low
```

```python
housing2 = long_lat.merge(cal_housing_med) #merge long_lat with cal_housing_med
```

```python
housing3 = long_lat.merge(cal_housing_high) #merge long_lat with cal_housing_high
```

```python
housing = pd.concat([housing1, housing2, housing3],ignore_index = True) # combine housing1,2 and 3 into 1 sheet with all day except Ocean Proximity
```

```python
housing = housing.merge(ocean_proximity, on = ['longitude', 'latitude']) #merge housing with ocean_proximity data
```

```python
housing = housing.sort_values('id', ascending = True) #sort DF by id in ascending order
```

```python
#housing['ocean_proximity'].value_counts()
housing['ocean_proximity'] = housing['ocean_proximity'].values.astype(str)
```

# Drop NA from medianHouseValue column

```python
housing['medianHouseValue'].isna().sum() #how many na values are in the medianHouseValue column
```

```python
housing = housing.dropna(subset = ['medianHouseValue']) #dropping na values from medianHouseValue column
```

# Fill NA values in housingMedianAge with Median value

```python
housing['housingMedianAge'].isna().sum() #how many NA values are in housingMedianAge column
```

```python
housing['housingMedianAge'].median() #what is the median housingMedianAge?
```

```python
housing['housingMedianAge'] = housing['housingMedianAge'].fillna(housing['housingMedianAge'].median()) #fill NA with median value
```

```python
# housing['housingMedianAge'].isna().sum() - resulted in 0 so all NA were filled
```

# Drop the state column from housing DF

```python
housing = housing.drop(columns = 'state')
```

# Change number format for medianIncome and medianHouseValue

```python
housing['medianIncome'] = housing['medianIncome'].div(10000).round(4) #scaling down medianIncome by a factor of 10000
```

```python
# establishing cap values at 15.0001 and .4999
housing.loc[housing['medianIncome'] <= .4999] = .4999
housing.loc[housing['medianIncome'] >= 15.0001] = 15.0001
```

```python
# revert medianHouseValue back to actual dollars
housing['medianHouseValue'] = housing['medianHouseValue']*1000
```

# Update Column names and order

```python
rename_dic = {'longitude' : 'longitude', 
          'latitude' : 'latitude',
          'housingMedianAge' : 'housing_median_age',
          'totalRooms' : 'total_rooms',
         'totalBedrooms' : 'total_bedrooms',
         'population' : 'population',
         'households' : 'households',
         'medianIncome' : 'median_income',
         'medianHouseValue' : 'median_house_value',
         'ocean_proximity' : 'ocean_proximity'}
housing.rename(columns=rename_dic, inplace = True)
```

```python
# reorder the columns
housing = housing[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population',
                'households','median_income', 'median_house_value', 'ocean_proximity']]
```

# Change all columns to float datatype except ocean_proximity

```python
housing[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 
         'median_income', 'median_house_value']] = housing[['longitude', 'latitude', 'housing_median_age', 'total_rooms', 
                                                            'total_bedrooms', 'population', 'households','median_income', 
                                                            'median_house_value']].values.astype(float)
```
